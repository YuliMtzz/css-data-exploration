```python
#Carga de librerias
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# carga de base de datos 
bookings = pd.read_csv("bookings_report_casas_sabro - Sheet1.csv")
bookings.head()
```

```python
#Limpieza y preparación de datos
bookings.info()
```

```python
# Eliminar columnas vacías
bookings = bookings.dropna(axis=1, how="all")
```

```python
#Confirmar que se eliminaron las columnas vacías
bookings.info()
```

```python
#Transformación de los datos
#Cambiar el data type de las columnas numericas y de las columnas que contienen fechas 

money_cols = ["Average Daily Rate", "Rental Revenue", "Total Revenue"]
for col in money_cols:
    bookings[col] = bookings[col].astype(str).str.replace(r"[^\d.]", "", regex=True).astype(float)

```

```python
#Cambiar el data type de las columnas que contienen fechas
date_cols = ["Check-in Date", "Check-out Date", "Booked Date"]
meses_es_en = {
    "ene": "Jan", "feb": "Feb", "mar": "Mar", "abr": "Apr",
    "may": "May", "jun": "Jun", "jul": "Jul", "ago": "Aug",
    "sep": "Sep", "oct": "Oct", "nov": "Nov", "dic": "Dec"
}
for col in date_cols:
    temp = bookings[col].astype(str).str.lower()
    for es, en in meses_es_en.items():
        temp = temp.str.replace(es, en, regex=False)
    bookings[col] = pd.to_datetime(temp, format="%d %b %Y", errors="coerce")

bookings.info()
```

```python
# Eliminar filas sin fecha válida ────────────────────
bookings = bookings.dropna(subset=["Check-in Date"]).copy()
print(f"Registros válidos: {len(bookings)} de 248")
```

```python
# Gaps entre reservas por propiedad
bookings = bookings.sort_values(["Listing ID", "Booked Date"])
bookings["prev_booked"]      = bookings.groupby("Listing ID")["Booked Date"].shift(1)
bookings["days_since_last"]  = (bookings["Booked Date"] - bookings["prev_booked"]).dt.days

# Flags
bookings["flag_last_minute"] = bookings["Booking Window (Days)"] <= 15
bookings["flag_orphan"]      = bookings["Length of Stay (Days)"].between(1, 9) & bookings["Booking Window (Days)"].between(0, 30)
bookings["flag_recency"]     = bookings["days_since_last"] >= 15
bookings["flag_discount"]    = bookings["flag_last_minute"] | bookings["flag_orphan"] | bookings["flag_recency"]

print(f"Total reservas: {len(bookings)}")
print(f"Con descuento:  {bookings['flag_discount'].sum()} ({bookings['flag_discount'].mean():.1%})")
```

```python
bookings["Año"]     = bookings["Check-in Date"].dt.year.astype(int)
bookings["Mes_num"] = bookings["Check-in Date"].dt.month
bookings["BW_calc"] = (bookings["Check-in Date"] - bookings["Booked Date"]).dt.days

bookings.info()
```

```python
#Suposición de lo que se fuese ganado si no se hubiesen aplicado descuentos
#Last Minute: 30% de descuento
#Orphan: 20% de descuento
#Recency: 10% de descuento (Promedio, entre 5% y 15%)
bookings = bookings.sort_values(["Listing ID", "Booked Date"])
bookings["prev_booked"]     = bookings.groupby("Listing ID")["Booked Date"].shift(1)
bookings["days_since_last"] = (bookings["Booked Date"] - bookings["prev_booked"]).dt.days

# Flags
bookings["flag_last_minute"] = bookings["Booking Window (Days)"] <= 15
bookings["flag_orphan"]      = bookings["Length of Stay (Days)"].between(1, 9) & bookings["Booking Window (Days)"].between(0, 30)
bookings["flag_recency"]     = bookings["days_since_last"] >= 30
bookings["flag_discount"]    = bookings["flag_last_minute"] | bookings["flag_orphan"] | bookings["flag_recency"]

print(f"Total reservas: {len(bookings)}")
print(f"Con descuento:  {bookings['flag_discount'].sum()} ({bookings['flag_discount'].mean():.1%})")
```

```python
#Validar que las flags funcionan correctamente
print("Last minute:", bookings["flag_last_minute"].sum())
print("Orphan:",      bookings["flag_orphan"].sum())
print("Recency:",     bookings["flag_recency"].sum())
print("Total con descuento:", bookings["flag_discount"].sum())
```

```python
print(bookings.columns.tolist())
```

```python
def adr_ideal(row):
    if row["flag_last_minute"]: return row["Average Daily Rate"] / 0.70
    if row["flag_orphan"]:      return row["Average Daily Rate"] / 0.80
    if row["flag_recency"]:     return row["Average Daily Rate"] / 0.90
    return row["Average Daily Rate"]

bookings["ADR_ideal"]       = bookings.apply(adr_ideal, axis=1)
bookings["Revenue_ideal"]   = bookings["ADR_ideal"] * bookings["Length of Stay (Days)"]
bookings["Revenue_perdido"] = bookings["Revenue_ideal"] - bookings["Total Revenue"]

# Verificar que se crearon
print(bookings[["Average Daily Rate", "ADR_ideal", "Revenue_ideal", "Revenue_perdido"]].head())
```

```python
#Ganancia total vs ganancia ideal
revenue_real     = bookings["Total Revenue"].sum()
revenue_ideal    = bookings["Revenue_ideal"].sum()
revenue_perdido  = revenue_ideal - revenue_real

print(f" Revenue real: ${revenue_real:,.0f}")
print(f" Revenue ideal: ${revenue_ideal:,.0f}")
print(f" Revenue lost: ${revenue_perdido:,.0f}")

# Porcentaje de ganancia perdida con descuento 
pct_perdido      = revenue_perdido / revenue_ideal
print(f" Lost: {pct_perdido:.1%}")
```

```python
#Ganancias por año
#En el año 2026 es hasta el mes de abril, mes en donde hacemos el analisis 
#Hay que tener en cuenta que las ganacias en ese año pueden ser menores por esto, y no necesariamente por una disminución en las reservas o en el precio promedio.
#Ademas, se debe tener en cuenta que desde finales de 2025 Casa vip dejo de operar sus 6 habiatciones 
bookings["Año"] = bookings["Check-in Date"].dt.year

# Resumen anual
anual = bookings.groupby("Año").agg(
    reservas        = ("Reservation ID", "count"),
    revenue_real    = ("Total Revenue", "sum"),
    revenue_ideal   = ("Revenue_ideal", "sum"),
).reset_index()

anual["revenue_perdido"] = anual["revenue_ideal"] - anual["revenue_real"]
anual["pct_perdido"]     = (anual["revenue_perdido"] / anual["revenue_ideal"] * 100).round(1)

# Formatear valores para lectura
anual_display = anual.copy()
for col in ["revenue_real", "revenue_ideal", "revenue_perdido"]:
    anual_display[col] = anual_display[col].map("${:,.0f}".format)
anual_display["pct_perdido"] = anual_display["pct_perdido"].map("{}%".format)

anual_display
```
